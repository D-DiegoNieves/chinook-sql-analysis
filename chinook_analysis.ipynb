```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load all CSVs
revenue_by_country = pd.read_csv('revenue_by_country.csv')
genre_sales = pd.read_csv('genre_sales.csv')
top_artists = pd.read_csv('top_artists.csv')
monthly_revenue = pd.read_csv('monthly_revenue.csv')
avg_order_value = pd.read_csv('avg_order_value.csv')

print("All files loaded successfully")
```

```python
# Set style
sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(3, 2, figsize=(16, 18))
fig.suptitle('Chinook Music Store - SQL Analysis', fontsize=20, fontweight='bold', y=1.01)

# Q1: Revenue by Country
axes[0, 0].barh(revenue_by_country['billing_country'], revenue_by_country['total_revenue'], color='steelblue')
axes[0, 0].set_title('Revenue by Country (Top 10)')
axes[0, 0].set_xlabel('Total Revenue ($)')
axes[0, 0].invert_yaxis()

# Q3: Genre Sales
axes[0, 1].barh(genre_sales['genre'], genre_sales['tracks_sold'], color='coral')
axes[0, 1].set_title('Tracks Sold by Genre (Top 10)')
axes[0, 1].set_xlabel('Tracks Sold')
axes[0, 1].invert_yaxis()

# Q4: Top Artists
axes[1, 0].barh(top_artists['artist'], top_artists['tracks_sold'], color='mediumseagreen')
axes[1, 0].set_title('Best Selling Artists (Top 10)')
axes[1, 0].set_xlabel('Tracks Sold')
axes[1, 0].invert_yaxis()

# Q7: Monthly Revenue Trend
monthly_revenue['month'] = pd.to_datetime(monthly_revenue['month'])
axes[1, 1].plot(monthly_revenue['month'], monthly_revenue['monthly_revenue'], color='steelblue', linewidth=2)
axes[1, 1].set_title('Monthly Revenue Trend')
axes[1, 1].set_xlabel('Month')
axes[1, 1].set_ylabel('Revenue ($)')
axes[1, 1].tick_params(axis='x', rotation=45)

# Q10: Avg Order Value by Country
axes[2, 0].barh(avg_order_value['billing_country'], avg_order_value['avg_order_value'], color='mediumpurple')
axes[2, 0].set_title('Avg Order Value by Country (Top 10)')
axes[2, 0].set_xlabel('Avg Order Value ($)')
axes[2, 0].invert_yaxis()

# Hide empty 6th subplot
axes[2, 1].set_visible(False)

plt.tight_layout()
plt.savefig('chinook_charts.png', dpi=150, bbox_inches='tight')
plt.show()
print("Charts saved as chinook_charts.png")
```
